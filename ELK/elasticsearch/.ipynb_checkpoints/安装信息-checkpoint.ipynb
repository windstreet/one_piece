# 如何找到您的elasticsearch主目录并且安装插件

## 1）找到您的主目录

`curl "localhost:9200/_nodes/settings?pretty=true"`  
    或者  
`brew info elasticsearch`

2）cd 到位置 ( 示例settings.path.home值：/usr/local/Cellar/elasticsearch/6.2.2 )


$ cd /usr/local/Cellar/elasticsearch/6.2.2 


3）安装插件(插件示例：mobz/elasticsearch-head ) 

$ bin/plugin -install mobz/elasticsearch-head
