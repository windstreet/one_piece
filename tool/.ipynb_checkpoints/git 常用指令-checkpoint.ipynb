# git 常用指令


## 1）查看分支

* 本地： git branch           
* 远程： git branch -a      
>前面带 * 号的代表你当前工作目录所处的分支  

## 2）在本地新建分支并切换到该分支

* git checkout -b spiders_migration 					                 
>spiders_migration 为本地分支名。

## 3）将新建的分支推送到远程仓库

* git push --set-upstream origin spiders_migration				
>origin: 远程主机名，spiders_migration: 远程分支名。以后推送就直接 git push 即可。

## 4）删除本地分支 

* git branch -d BranchName

## 5）将远程的 origin/feature 分支与 本地的 feature 分支关联起来

* git branch --set-upstream-to=origin/feature feature       

## 6）打印你的每一次操作记录

* git reflogq

  输出：
   >c7edbfe HEAD@{0}: reset: moving to c7edbfefab1bdbef6cb60d2a7bb97aa80f022687
	470e9c2 HEAD@{1}: reset: moving to 470e9c2

* git reset --hard b45959e		
   >找到你操作的id如：b45959e，就可以回退到这个版本

## 7）添加快捷键

* vim ~/.zshrc

```
alias gs="git status"
alias gd="git add"
alias gc="git commit -m"
alias gph="git push"
alias gpl="git pull"
```

* source ~/.zshrc

## 8）合并分支

* git merge --no-ff feature/0402_sync_stock_transfer_order_arrival
>当前所在分支为 master

## 9）Git flow

* git flow init 
* git flow hotfix finish -n 04_30_fix_purchase_order_item       
>将 hotfix 分支合起来，当前所在分支即为 04_30_fix_purchase_order_item。


## 10）撤销

* 撤销所有的已经add的文件：git reset HEAD .		
>有个点别漏了，'HEAD .'

* 如果是撤销某个文件或文件夹：git reset HEAD -filename

## 11）清理了部分远程分支后，本地同步一下

* git remote update origin --prune
