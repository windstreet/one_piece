# 项目初始化


## 1）环境与依赖

`conda create --name overlord python=3.7.1`  
`pip install -r requirements.txt`  
`pip install -r libs/goblin/requirements.txt`  # 已弃用

## 2）运行

### 2.0）我的补充：
第一次拉要pull一下docker镜像:  
`docker pull ubuntu:16.04`

检查本地的镜像：  
`docker images`

### 2.1）创建数据库
`createdb overlord`  

删除数据库：  
`dropdb overlord`  

### 2.2）数据迁移
每次有数据库更新时都要执行一遍，然后再migrate，正式版未发布前不要提交上来。  

`python manage.py makemigrations`   
`python manage.py migrate`  

### 2.3）创建超级用户
`python manage.py createsuperuser`

补充：postgresql  创建用户，删除用户 
`createuser --superuser postgres`
`dropuser postgres`

### 2.4）启动服务
服务1）`python manage.py runserver`

    服务地址：http://localhost:8000

在另一个shell里启用celery服务  
服务2）`celery -A overlord worker`
补充：
`celery -A overlord.celery_app worker` # 前者无效时可以尝试    

服务3）`celery -A overlord beat`

### 2.5）其他配置

* 如果本地的 `Docker` 不是 `swarm` 模式，需要初始化：`docker swarm init`

* 如果本地的 mq 未安装 `rabbitmq_delayed_message_exchange` 插件，请自行安装。

## Goblin 项目引入说明

在 `overlord` 的环境下，切到 `goblin` 主目录，然后运行 `python setup.py install`
